# AI-Powered Campaign Performance Analyzer

## Overview
This agent automatically analyzes B2B marketing campaign results, compares performance 
against industry benchmarks, flags anomalies, and generates actionable recommendations 
— replacing hours of manual reporting with instant AI-driven insights.

## Analysis Framework
For each campaign the agent produces:
1. **Performance Rating** — Excellent / Good / Needs Improvement / Critical
2. **Key Metrics Analysis** — open rate, click rate, cost per MQL, pipeline ROI
3. **Benchmark Comparison** — actual vs target with variance
4. **Anomaly Detection** — unusual patterns that indicate problems
5. **Recommendations** — immediate, short-term, and strategic actions
6. **Executive Summary** — 3-sentence CMO-ready summary

## Real-World Application
In a production MarTech stack this agent would:
1. Pull campaign data automatically from Marketo or Salesforce reports via API
2. Run analysis after every campaign send
3. Post summary to Slack or email marketing team automatically
4. Flag underperforming campaigns before weekly review meetings
5. Generate CMO-ready reporting in seconds instead of hours

```python
from dotenv import load_dotenv 
import os
import anthropic

load_dotenv()
client = anthropic.Anthropic()
```

```python
campaigns = [
    {
        "name": "Q1 Enterprise SaaS Email Campaign",
        "type": "Email",
        "segment": "Enterprise VP+",
        "sends": 5000,
        "opens": 900,
        "clicks": 180,
        "unsubscribes": 45,
        "mqls_generated": 12,
        "pipeline_generated": 85000,
        "cost": 2500,
        "benchmark_open_rate": 0.25,
        "benchmark_click_rate": 0.05
    },
    {
        "name": "APAC Webinar - Marketing Automation",
        "type": "Webinar",
        "segment": "Mid-Market APAC",
        "sends": 2000,
        "opens": 680,
        "clicks": 320,
        "unsubscribes": 8,
        "mqls_generated": 28,
        "pipeline_generated": 120000,
        "cost": 5000,
        "benchmark_open_rate": 0.35,
        "benchmark_click_rate": 0.15
    },
    {
        "name": "SMB Nurture Stream Q1",
        "type": "Nurture",
        "segment": "SMB All Regions",
        "sends": 8000,
        "opens": 960,
        "clicks": 160,
        "unsubscribes": 220,
        "mqls_generated": 8,
        "pipeline_generated": 15000,
        "cost": 1200,
        "benchmark_open_rate": 0.22,
        "benchmark_click_rate": 0.04
    }
]
```

```python
analysis_prompt = """
You are a B2B SaaS marketing performance expert. Analyze these campaign results and provide:

1. PERFORMANCE RATING: Score each campaign (Excellent/Good/Needs Improvement/Critical)
2. KEY METRICS ANALYSIS: Calculate and interpret open rate, click rate, cost per MQL, pipeline ROI
3. BENCHMARK COMPARISON: Compare against provided benchmarks
4. ANOMALIES: Flag anything unusual or concerning
5. RECOMMENDATIONS: Specific actionable next steps for each campaign
6. EXECUTIVE SUMMARY: 3-sentence summary a CMO can read in 30 seconds

Format each campaign analysis clearly with the campaign name as header.
"""
```

```python
for campaign in campaigns:
    print(f"\n{'='*60}")
    print(f"Analyzing: {campaign ['name']}")
    print(f"\n{"="*60}")

    message = client.messages.create(
        model ='claude-haiku-4-5-20251001',
        max_tokens = 2048,
        messages = [
            {
                "role":"user",
                "content": f"{analysis_prompt}\n\nAnalyze this campaign:\n{campaign}"
            }
        ]
    )
    print (message.content[0].text)
```
